Patient Class

```python
class Patient:
  def __init__(self,id,name,age,bloodgroup,prev=None,next=None):
   self.id=id
   self.name=name
   self.age=age
   self.bg=bloodgroup
   self.prev=prev
   self.next=next
```

WRM (waiting room management) class

```python
class WRM:
  def __init__(self):
    self.dh=Patient(None,None,None,None) # initializing with only one node
    self.dh.prev = self.dh
    self.dh.next = self.dh    # prev and next are DH itself, all other attributes as None

  def RegisterPatient(self, id, name, age, bloodgroup):
    print('executing RegisterPatient()...')
    formr_last=self.dh.prev
    new_patient=Patient(id,name,age,bloodgroup,formr_last,self.dh)
    self.dh.prev=new_patient
    formr_last.next=new_patient
    print('Success registering patient')

  def ServePatient(self):
    to_serve=self.dh.next
    if self.dh.next==self.dh:
      print('No patient to be served')
    elif to_serve.next==self.dh:
      self.dh.prev=self.dh.next=self.dh
      print(f'{to_serve.name} is served')
    else:
      pending=to_serve.next
      pending.prev=self.dh
      self.dh.next=pending
      to_serve.prev=to_serve.next=None
      print(f'{to_serve.name} is served')
      to_serve.name=to_serve=None

  def CancelAll(self):
    if self.dh.next==self.dh:
      print('No patient to be served')
    else:
      self.dh.prev=self.dh.next=self.dh

  def CanDoctorGoHome(self):
    if self.dh.next==self.dh and self.dh.prev==self.dh:
       print('Yes')
    else:
      print('No')

  def ShowAllPatient(self):
    if self.dh.next==self.dh and self.dh.prev==self.dh:
       print('No patient')
    else:
      inode=self.dh.next
      while inode!=self.dh:
        print(inode.name)
        inode=inode.next

  def ReverseTheLine(self):
    inode=self.dh.next
    self.dh.prev,self.dh.next=self.dh.next,self.dh.prev

    while inode!=self.dh:
      inext=inode.next
      inode.prev,inode.next=inode.next,inode.prev
      inode=inext
```

Driver Code

```python
wrm = WRM()

while True:
    print("==Choose an Option==")
    print("1. RegisterPatient()")
    print("2. ServePatient()")
    print("3. CancelAll()")
    print("4. CanDoctorGoHome()")
    print("5. ShowAllPatient()")
    print("6. ReverseTheLine()")
    print("7. exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        id = input("Enter ID: ")
        name = input("Enter Name: ")
        age = input("Enter Age: ")
        bloodgroup = input("Enter Blood group: ")
        wrm.RegisterPatient(id, name, age, bloodgroup)
    elif choice == '2':
        wrm.ServePatient()
    elif choice == '3':
        wrm.CancelAll()
    elif choice == '4':
        wrm.CanDoctorGoHome()
    elif choice == '5':
        wrm.ShowAllPatient()
    elif choice == '6':
        wrm.ReverseTheLine()
    elif choice == '7':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```
