```python
class Patient:

  def __init__(self, id = None, name = None, age = None, bloodGroup = None, next = None, prev = None):

    self.id = id

    self.name = name

    self.age = age

    self.bloodGroup = bloodGroup

    self.next = next

    self.prev = prev

#=========================================================================================
```

```python
class WRM:

  def __init__(self):

    self.dh = Patient()

    self.dh.next = self.dh

    self.dh.prev = self.dh

  def registerPatient(self, id, name, age, bloodGroup):

    last_patient = self.dh.prev

    new_patient = Patient(id, name, age, bloodGroup, self.dh, last_patient)

    self.dh.prev = new_patient

    last_patient.next = new_patient

    print('Patient Successfully Registered')

  def servePatient(self):

    c = 0

    n = self.dh.next

    while(n != self.dh):

      c += 1

      n = n.next

    if(c == 0):

      print('No Patients Left')

    else:

      s_patient = self.dh.next

      n_patient = s_patient.next

      self.dh.next = n_patient

      n_patient.prev = self.dh

      print(f'{s_patient.name} is served.')

      s_patient = None

  def cancelAll(self):

    c = 0

    n = self.dh.next

    while(n != self.dh):

      c += 1

      n = n.next

    if(c == 0):

      print('No Patients Left')

    else:

      self.dh.next = self.dh

      self.dh.prev = self.dh

      print('All Appointments Cancelled.')

  def canDoctorGoHome(self):

    c = 0

    n = self.dh.next

    while(n != self.dh):

      c += 1

      n = n.next

    if(c == 0):

      return 'YES'

    else:

      return 'NO'

  def showAllPatient(self):

    if(self.dh.next == self.dh and self.dh.prev == self.dh):

      print('There is no patient in the waiting room.')

    else:

      n = self.dh.next

      c = 1

      while(n != self.dh):

        print(f'{c}. {n.name}')

        c += 1

        n = n.next

  def reverseTheLine(self):

    if(self.dh.next == self.dh and self.dh.prev == self.dh):

      print('There is no patient in the waiting room.')

    else:

      last_node = self.dh

      current_node = self.dh.next

      while(current_node != self.dh):

        next_node = current_node.next

        current_node.next, current_node.prev = current_node.prev, next_node

        current_node = next_node

      self.dh.next, self.dh.prev = self.dh.prev, last_node

      print('Reverse Successful.')
```

```python
# Driver Code

import numpy as np

dummyPatient = WRM()

print('***Welcome to Waiting Room Management (WRM) System***')

print('===Choose an Option')
print('1. RegisterPatient()')
print('2. ServePatient()')
print('3. CancelAll()')
print('4. CanDoctorGoHome()')
print('5. ShowAllPatient()')
print('6. ReverseTheLine()')
print('7. Exit')
print('============================\n')

a = int(input('Enter Your Choice: '))

while(a != 7):

  if(a == 1):

    print('executing RegisterPatient()...')

    dummyPatient.registerPatient(input('Enter Patient ID: '), input('Enter Patient Name: '), input('Enter Patient Age: '), input('Enter Patient Blood Group:'))

    print()

  elif(a == 2):

    print('executing ServePatient()...')

    dummyPatient.servePatient()

  elif(a == 3):

    print('executing CancelAll()...')

    dummyPatient.cancelAll()

  elif(a == 4):

    print('executing CanDoctorGoHome()...')

    print(dummyPatient.canDoctorGoHome())

  elif(a == 5):

    print('executing ShowAllPatients()...')

    print('Showing All Patients:')

    dummyPatient.showAllPatient()

  elif(a == 6):

    print('executing ReverseTheLine()...')

    dummyPatient.reverseTheLine()

  elif(a == 7):

    print('Exiting Program')

    break

  else:

    print('Invalid Input!')

    continue

  print('===Choose an Option')
  print('1. RegisterPatient()')
  print('2. ServePatient()')
  print('3. CancelAll()')
  print('4. CanDoctorGoHome()')
  print('5. ShowAllPatient()')
  print('6. ReverseTheLine()')
  print('7. Exit')

  a = int(input())
```
