```python
class Patient:
    def __init__(self, id=None, name=None, age=None, bloodgroup=None):
        self.id = id
        self.name = name
        self.age = age
        self.bloodgroup = bloodgroup
        self.next = None
        self.prev = None


class WRM:
    def __init__(self):
        self.head = Patient()
        self.head.next = self.head
        self.head.prev = self.head

    def RegisterPatient(self, id, name, age, bloodgroup):
        if not (id and name and age and bloodgroup):
            print('Patient cannot be added. Missing information.')
            return
        if bloodgroup not in ['B+', 'B-', 'A+', 'A-', 'O+', 'O-', 'AB+', 'AB-']:
            print('Patient cannot be added. Invalid blood group.')
            return

        new_patient = Patient(id, name, age, bloodgroup)
        old_patient = self.head.prev

        new_patient.next = self.head
        new_patient.prev = old_patient
        old_patient.next = new_patient
        self.head.prev = new_patient

    def ServePatient(self):
        first = self.head.next
        if first == self.head:
            print('No patient to serve')
            return None
        else:
            print(f"Serving patient: {first.name}")
            next_patient = first.next 
            prev_patient = first.prev 
            next_patient.prev = prev_patient 
            prev_patient.next = next_patient
            return first

    def CancelAll(self):
        self.head.next = self.head
        self.head.prev = self.head

    def CanDoctorGoHome(self):
        if self.head.next == self.head:
            return True
        else:
            return False

    def ShowAllPatient(self):
        present_patient = self.head.next
        if present_patient == self.head:
            print('There are no patients in the waiting area')
        while present_patient != self.head:
            print(
                f"ID: {present_patient.id}, Name: {present_patient.name}, Age: {present_patient.age}, Blood Group: {present_patient.bloodgroup}")
            present_patient = present_patient.next

    def ReverseTheLine(self):
        fast = self.head.prev
        last = self.head.next
        while fast != self.head:
            fast.next, fast.prev, last = last, fast.next, fast #Not update fast
        self.head.prev, last.next = last, self.head.prev
        self.head.next = last


wrm = WRM()

while True:
    print("Options:")
    print("a. Add Patient")
    print("b. Serve Patient")
    print("c. Show All patients")
    print("d. Can Doctor go Home?")
    print("e. Cancel all Appointment")
    print("f. ReverseTheLine")
    option = input("Enter option: ")

    if option == 'a':
        id = int(input("Enter id: "))
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        bloodgroup = input("Enter blood group: ")
        wrm.RegisterPatient(id, name, age, bloodgroup)
    elif option == 'b':
        wrm.ServePatient()
    elif option == 'c':
        wrm.ShowAllPatient()
    elif option == 'd':
        if wrm.CanDoctorGoHome():
            print("Yes, doctor can go home")
        else:
            print("No, doctor cannot go home")
    elif option == 'e':
        wrm.CancelAll()
        print("Successfully canceled all appointments")
    elif option == 'f':
        wrm.ReverseTheLine()
    else:
        print('Command is not accepted')
        break
```
