```python
import numpy as np
class Patient:
    def __init__(self, id, name, age, blood):
        self.id = id
        self.name = name
        self.age = age
        self.blood = blood
        self.next = None
        self.prev = None

class WRM:
    def __init__(self):
        self.head = None
        self.tail = None

    def RegisterPatient(self, id, name, age, blood):
       new = Patient(id, name, age, blood)
       if self.head is None:
            self.head = new
            self.tail = new
            new.next = new
            new.prev = new
       else:
            new.next = self.head
            new.prev = self.tail
            self.tail.next = new
            self.head.prev = new
            self.tail = new

    def ServePatient(self):
        if self.head is not None:
            checked = self.head
            if self.head == self.head.next:
                self.head = None
                self.tail = None
            else:
                self.head = self.head.next
                self.head.prev = self.tail
                self.tail.next = self.head
            return checked.name
        else:
            return "No patients in the waiting room."

    def CancelAll(self):
        self.head = None
        self.tail = None

    def CanDoctorGoHome(self):
        return self.head is None

    def ShowAllPatient(self):
        if self.head is not None:
            current = self.head
            while True:
                print(current.name)
                current = current.next
                if current == self.head:
                    break
        else:
            print("No patients in the waiting room.")

    def ReverseTheLine(self):
        if self.head is not None:
            self.head, self.tail = self.tail, self.head
            current = self.head
            while True:
                current.prev, current.next = current.next, current.prev
                current = current.next #This should be current.prev
                if current == self.head:
                    break
        else:
            print("No patients in the waiting room.")





# Tester code
wrm = WRM()

while True:
    print("\nOptions:")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Reverse The Line")
    print("7. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood = input("Enter patient blood group: ")
        wrm.RegisterPatient(id, name, age, blood)
        print("Patient added successfully.")

    elif choice == 2:
        checked = wrm.ServePatient()
        print(f"Serving patient: {checked}")

    elif choice == 3:
        wrm.ShowAllPatient()

    elif choice == 4:
        if wrm.CanDoctorGoHome():
            print("Yes, the doctor can go home.")
        else:
            print("No, there are patients waiting.")

    elif choice == 5:
        wrm.CancelAll()
        print("All appointments canceled.")

    elif choice == 6:
        wrm.ReverseTheLine()
        print("Line order reversed.")

    elif choice == 7:
        break

    else:
        print("Invalid choice. Please choose a valid option.")
```
