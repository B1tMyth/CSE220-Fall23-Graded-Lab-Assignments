```python
class Patient:
    def __init__(self,id,name,age,bg,next,prev):
        self.id=id
        self.name=name
        self.age=age
        self.bg=bg
        self.next=next
        self.prev=prev

class WRM:
    def __init__(self):
        dh=Patient(None,None,None,None,None,None)
        dh.next=dh
        dh.prev=dh
        self.dh=dh

    def RegisterPatient(self,id,name,age,bg):
        last=self.dh.prev
        new=Patient(id,name,age,bg,self.dh,last)
        last.next=new
        self.dh.prev=new
        print("Registration Successful")
        print(f"Patient Details:\nId: {id}, Name: {name}, Age: {age}, Blood Group: {bg}")

    def ServePatient(self):
        servedP=self.dh.next
        if servedP==self.dh:
            print("No Patient To Be Served")
            return
        nextP=servedP.next
        self.dh.next=nextP
        nextP.prev=self.dh
        servedP.next,servedP.prev=None,None
        print("Served Patient:",servedP.name)
        servedP.id,servedP.name,servedP.age,servedP.bg=None,None,None,None

    def CancelAll(self):
        if self.dh.next==self.dh:
            print("No Patient in WRM")
        else:
            self.dh.next=self.dh
            self.dh.prev=self.dh
            print("All Appoinments are Cancelled")

    def CanDoctorGoHome(self):
        if self.dh.next==self.dh:
            print("Yes")
        else:
            print("No")

    def ShowAllPatient(self):
        tail=self.dh.next
        if tail==self.dh:
            print("No Patient Left")
        else:
            print("Patient Names: ")
            while tail.next!=self.dh:
                print(tail.name,end=", ")
                tail=tail.next
            print(tail.name)

    def ReverseTheLine(self):
        if self.dh.next==self.dh:
            print("Line can't be reversed as there's no patient available")
        else:
            current=self.dh.next
            while current.next!=self.dh:
                after=current.next
                current.next=current.prev
                current.prev=after
                current=current.prev
            current.next=current.prev
            current.prev=self.dh
            temp=self.dh.next
            self.dh.next=current
            self.dh.prev=temp
            print("The Line is Reversed")

            self.ShowAllPatient()


#DRIVER CODE
w=WRM()
print("*"*50,"Welcome To WRM","*"*50)
while True:
    command=int(input("\nChoose Your Option: \n1.Register Patient\n2.See Patients\n3.Served Patient\n4.Cancel Info\n5.Doctor In Home?\n6.Reversed Line\n0.Exit\n"))
    if command==0:
        print("Finished")
        print("!"*100)
        break
    elif command==1:
        print("Register Your Patient Now!")
        inp1=int(input("Enter Your ID: "))
        inp2=input("Enter Your Name: ")
        inp3=int(input("Enter Your Age: "))
        inp4=input("Enter Your Blood Group: ")
        w.RegisterPatient(inp1,inp2,inp3,inp4)
        print("="*100)
    elif command==2:
        print("Patients Status:")
        w.ShowAllPatient()
        print("="*100)
    elif command==3:
        print("Service Status:")
        w.ServePatient()
        print("="*100)
    elif command==4:
        print("Cancellation Status:")
        w.CancelAll()
        print("="*100)
    elif command==5:
        print("Can Doctor Go Home?",end=": ")
        w.CanDoctorGoHome()
        print("="*100)
    elif command==6:
        print("Reversing The Line.......")
        w.ReverseTheLine()
        print("="*100)
    else:
        print("Invalid Option, Try Again")
        print("="*100)
```
