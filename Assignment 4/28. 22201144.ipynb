```python
class Patient:
    idx = -1
    def __init__(self, idn = None, name = None, age = None, bg = None):
        Patient.idx += 1
        self.idn = idn
        self.name = name
        self.age = age
        self.bg = bg

        self.next = None
        self.prev = None



class WRM:
    def __init__(self):
        self.dh = Patient()
        self.dh.next = self.dh
        self.dh.prev = self.dh
        self.tail = self.dh

    def registerPatient(self, idn, name, age, bloodgroup):
        patient = Patient(idn, name, age, bloodgroup)
        self.tail.next = patient
        patient.next = self.dh
        patient.prev = self.tail
        self.dh.prev = patient
        self.tail = patient

    def servePatient(self):
        pat = self.dh.next
        next_serve = pat.next
        self.dh.next = next_serve
        next_serve.prev = self.dh

        pat.next = None
        pat.prev = None

        return pat.name


    def cancelAll(self):
        self.dh.next = self.dh
        self.dh.prev = self.dh
        return True

    def canDoctorGoHome(self):
        if self.dh.next == self.dh:
            return True
        else:
            return False

    def showAllPatients(self):
        temp = self.dh.next
        while temp != self.dh:
            print(temp.name)
            temp = temp.next

    def reverseTheLine(self):
        n = self.dh
        tail = n.prev
        temp = n.next
        while n.prev != temp and n.next != tail:

            n.next = tail
            n.prev = temp
            tail = n
            temp = temp.next
            n = n.prev

        return True

wrm = WRM()

while True:
    print("\nOptions:")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Reverse the Line")
    print("7. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        bloodgroup = input("Enter patient blood group: ")
        wrm.registerPatient(id, name, age, bloodgroup)
        print("Patient added successfully!")

    elif choice == 2:
        served_patient = wrm.servePatient()
        if served_patient:
            print("Serving patient:", served_patient.name)
        else:
            print("No patients to serve.")

    elif choice == 3:
        print("All patients in line:")
        wrm.showAllPatients()

    elif choice == 4:
        if wrm.canDoctorGoHome():
            print("Yes,Doctor can go home.")
        else:
            print("No,Doctor cannot go home.")

    elif choice == 5:
        wrm.cancelAll()
        print("All appointments canceled.")

    elif choice == 6:
        wrm.reverseTheLine()
        print("Line order reversed.")

    elif choice == 7:
        break

```
