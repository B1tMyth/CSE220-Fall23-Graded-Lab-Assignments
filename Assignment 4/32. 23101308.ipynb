```python
class Patient:
  def __init__(self,id=None,name=None,age=None,blood_group=None,next=None,prev=None):
    self.id=id
    self.name=name
    self.age=age
    self.group=blood_group
    self.next=next
    self.prev=prev
```

```python
class WRM:
  def __init__(self):
    self.head=Patient()
    self.head.next=self.head
    self.head.prev=self.head
    self.last_patient = self.head

  def RegisterPatient(self,id, name, age, bloodgroup):
    new_patient=Patient(id, name, age, bloodgroup,self.head,self.last_patient)
    self.last_patient.next=new_patient
    self.last_patient=self.last_patient.next
    self.head.prev=self.last_patient
    print("Patient Registered Successfully")


  def ServePatient(self):
    if self.head.next==self.head:
      print("no patient")
    else:
      serve=self.head.next
      next_patient=serve.next
      self.head.next=next_patient
      next_patient.prev=self.head
      print(f"{serve.name} served successfully")
      serve=None

  def CancelAll(self):
    if self.head.next==self.head:
      print("No Patient to cancel")
    else:
      self.head.next=self.head
      self.head.prev=self.head
      print("Cancelled all patient")

  def CanDoctorGoHome(self):
    if self.head.next==self.head:
      print("yes")
    else:
      print("NO")

  def ShowAllPatient(self):
    temp=self.head.next
    while temp!=self.head:
      print(temp.name)
      temp=temp.next

  def ReverseTheLine(self):
    current=self.head.next
    if self.head.next == self.head:
      print("Reverse failed")
    else:
      while current!=self.head:
        save = current.prev
        current.prev = current.next
        current.next = save
        current = current.prev
      save = current.prev
      current.prev = current.next
      current.next = save
      print("Reverse successful")


waiting_room = WRM()
while True:
  print("1. Register a patient")
  print("2. Serve a patient")
  print("3. Show all patients")
  print("4. Can doctor go home?")
  print("5. Cancel all appointments")
  print("6. Reverse the line")
  print("7. Stop")
  choice = int(input("Enter your choice: "))

  if choice == 1:
    id = int(input("Enter patient ID: "))
    name = input("Enter patient name: ")
    age = int(input("Enter patient age: "))
    blood_group = input("Enter patient blood group: ")
    waiting_room.RegisterPatient(id, name, age, blood_group)

  elif choice == 2:
    waiting_room.ServePatient()

  elif choice == 3:
    waiting_room.ShowAllPatient()

  elif choice == 4:
    waiting_room.CanDoctorGoHome()

  elif choice == 5:
    waiting_room.CancelAll()

  elif choice == 6:
    waiting_room.ReverseTheLine()
  elif choice==7:
    break
print()
```
