```python
class Patient:
  def __init__(self, id, name, age, blood_group):
    self.id=id
    self.name=name
    self.age=age
    self.blood_group=blood_group
```

```python
class DoublyNode:
    def __init__(self, elem=None, next=None, prev=None):
        self.elem = elem
        self.next = next
        self.prev = prev
```

```python
class WRM:
  def __init__(self):
    self.head = DoublyNode()
    self.head.next = self.head
    self.head.prev = self.head

  def RegisterPatient(self, id, name, age, bloodgroup):
    store = Patient(id, name, age, blood_group)
    new = DoublyNode(store)

    if self.head != None:
      val= self.head.prev
      new.next=self.head
      new.prev=val
      self.head.prev=new
      val.next=new
    else:
      self.head = new    # initializing new as new head
      new.next = new
      new.prev = new

  def ServePatient(self):
    value= self.head.next
    if self.head.next == self.head:
      return None

    self.head.next = value.next
    if value.next == None:
      value.prev = None
      value.next = None
    else:
      value.next.prev= self.head
      value.prev = None
      value.next = None

    return value.elem

  def CancelAll(self):
    if self.head.next != self.head:
      self.head.next= self.head
      self.head.prev= self.head
    else:
      print("No appointments available")

  def CanDoctorGoHome(self):
    if self.head.next != self.head:
      return False
    else:
      return True

  def ShowAllPatient(self):
    h= self.head.next
    if h == self.head:
      print(f"No Patients available")

    while h != self.head:
      print(f"PATIENT NAME: {h.elem.name}")
      h= h.next


  def ReverseTheLine(self):
    if self.head.next == self.head:
      print("Can't reverse, not enough patient\n")
    else:
      h=self.head
      t=self.head.prev
      while h.next.next != self.head:
        node = t.prev
        t.next = h.next
        t.prev= h
        h.next= t
        t.next.prev=t
        h=t
        t=node
        t.next=self.head
        self.head.prev=t
      print("Reversed the sequence of the patients successfully!")



patient_type = WRM()

while True:
  print(f"SERVICE OPTIONS")
  print(f"1. Add Patient\n2. Serve Patient\n3. Cancel all Appointment\
  \n4. Can Doctor go Home?\n5. Show All patients\n6. ReverseTheLine\n")

  select_option = int(input("OPTION NO: "))

  if select_option == 1:
    id= int(input('PATIENT ID: '))
    name= input('PATIENT NAME: ')
    age= int(input('PATIENT AGE: '))
    blood_group= input('PATIENT BLOOD GROUP: ')
    patient_type.RegisterPatient(id,name,age,blood_group)

    print("Successfully Added\n")

  elif select_option == 2:
    p= patient_type.ServePatient()
    if p != None:
      print(f'Providing services to: {p.name}')
    else:
      print(f'No Patient was Found\n')

  elif select_option == 3:
    patient_type.CancelAll()
    print('Appointment Canceled\n')

  elif select_option == 4:
    flag_value= patient_type.CanDoctorGoHome()
    if flag_value == True:
      print("Doctor Can Go Home\n")
    else:
      print("Doctor Can't Go Home\n")

  elif select_option == 5:
    patient_type.ShowAllPatient()

  elif select_option == 6:
    patient_type.ReverseTheLine()

  else:
    print("INVALID OPTION!\nTRY AGAIN")
    break
```
