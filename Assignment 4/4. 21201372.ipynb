```python
class Patient:
    def __init__(self,id,name,age,bloodtype):
        self.id=id
        self.name=name
        self.age=age
        self.bloodtype=bloodtype
        self.prev=None
        self.next=None
class WRM:
    def __init__(self):
        self.head = None
        self.tail = None

    def registerPatient(self, id, name, age, bloodgroup):
        patient = Patient(id, name, age, bloodgroup)
        if self.head is None:
          self.head = patient
          self.tail = patient
          patient.next = patient
          patient.prev = patient
        else:
          patient.prev = self.tail
          patient.next = self.head
          self.head.prev = patient
          self.tail.next = patient
          self.tail = patient
    def SrvPatient(self):
        if self.head is None:
            print("NO patient")
            return
        crnt_patient=self.head
        if self.head==self.tail:
            self.head=None
            self.tail=None
        else:
            self.head=self.head.next
            self.tail.next=self.head
            self.head.prev=self.tail
        print("Serving:",crnt_patient.name)
    def cancelAll(self):
        if self.head is None:
            print("NO PATIENT")
            return
        self.head=None
        self.tail=None
        print("All appointments canceled")
    def canDoctorGoHome(self):
        if self.head is None:
            return True
        else:
            return False
    def showAllPatient(self):
        if self.head is None:
            print("NO patient in room")
            return
        crnt_patient=self.head
        print("Waiting....")
        while True:
            print("ID:",crnt_patient.id)
            if crnt_patient==self.tail:
                break
            crnt_patient=crnt_patient.next
    def reverseTheLine(self):
        if self.head is None:
            print("No patients in the waiting room.")
            return
        current_patient = self.head
        while True:
            current_patient.prev, current_patient.next = current_patient.next, current_patient.prev
            current_patient = current_patient.next
            if current_patient == self.head:
                break
        self.head, self.tail = self.tail, self.head

def displayOptions():
  print("\n----- Waiting Room Management System Menu -----")
  print("1. Add Patient")
  print("2. Serve Patient")
  print("3. Show All Patients")
  print("4. Can Doctor Go Home?")
  print("5. Cancel All Appointments")
  print("6. The patient line has been reversed")
  print("7. Exit")

system = WRM()

while True:
  displayOptions()
  choice = int(input("Enter your choice: "))

  if choice == 1:
    id = int(input("Enter patient ID: "))
    name = input("Enter patient name: ")
    age = int(input("Enter patient age: "))
    bloodgroup = input("Enter patient blood group: ")
    system.registerPatient(id, name, age, bloodgroup)
    print("Patient added successfully.")

  elif choice == 2:
    system.SrvPatient()

  elif choice == 3:
    system.showAllPatient()

  elif choice == 4:
    if system.canDoctorGoHome():
      print("Yes, doctor can go home.")
    else:
      print("No, patients are waiting.")

  elif choice ==5:
    system.cancelAll()

  elif choice == 6:
    system.reverseTheLine()
  elif choice == 7:
    print("Exiting...")
    break
```
