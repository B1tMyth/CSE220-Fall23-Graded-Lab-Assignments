```python
class Patient:
  def __init__(self,id,name,age,blood):
    self.id=id
    self.name=name
    self.age=age
    self.blood=blood

class DoublyNode:
  def __init__(self,elem,next,prev):
    self.elem=elem
    self.next=next
    self.prev=prev

class WRM:
  def __init__(self):
    self.dh=DoublyNode(None,None,None)
    self.dh.next=self.dh
    self.dh.prev=self.dh
    self.tail=self.dh

  def RegisterPatient(self,id,name,age,blood):
    patient= Patient(id,name,age,blood)
    n=DoublyNode(patient,self.dh,self.tail)
    self.tail.next=n
    self.tail=self.tail.next
    self.dh.prev=self.tail
    print("Registration Successfull")

  def servePatient(self):
    if self.dh.next==self.dh:
      print("No patient to be served")
    else:
      first_patient=self.dh.next
      next_patient=first_patient.next
      next_patient.prev=self.dh
      self.dh.next=next_patient
      print(first_patient.elem.name,"is served")

  def CancelAll(self):
    self.dh.next=self.dh #Forgot to update self.dh.prev
    print("All appointments are canceled. Doctor can go to lunch.")

  def CanDoctorGoHome(self):
    if self.dh.next==self.dh:
      print("Yes")
    else:
      print("No")

  def ShowAllPatient(self):
    if self.dh.next==self.dh:
      print("No patient in the waiting the list")
    else:
      temp=self.dh.next
      while temp!=self.dh:
        print(temp.elem.id,end=",")
        temp=temp.next
      print()

  def reverse_the_line(self):
      current_node = self.tail
      while current_node != self.dh:
         print(current_node.elem.name)
         current_node = current_node.prev

wrm=WRM()
user_input=0
while user_input!=6:
  print()
  print("==Choose an Option==")
  print("Options:")
  print("1. Add Patient")
  print("2. Serve Patient")
  print("3. Show All Patients")
  print("4. Can Doctor Go Home?")
  print("5. Cancel All Appointments")
  print("6. Reverse The Line")
  print("7. Exit")
  print()
  user_input=int(input("Enter your option:"))
  if user_input==1:
    id=int(input("Enter ID: "))
    name=str(input("Enter Name: "))
    age=int(input("Enter Age: "))
    blood=str(input("Enter Blood Group: "))
    wrm.RegisterPatient(id,name,age,blood)
  elif user_input==2:
    wrm.servePatient()
  elif user_input==3:
    wrm.CancelAll()
  elif user_input==4:
    wrm.CanDoctorGoHome()
  elif user_input==5:
    wrm.ShowAllPatient()
  elif user_input==6:
    wrm.reverse_the_line()
  elif user_input==7:
    print("Thank you.Have a good day.")
  else:
    print("Invalid Number")
```
