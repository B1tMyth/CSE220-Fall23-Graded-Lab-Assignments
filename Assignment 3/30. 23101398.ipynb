```python
import numpy as np

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def find_right_rotations(self):
        if not self.head:
            return 0

        current = self.head
        rotations = 0

        while current and current.next:
            if current.data > current.next.data:
                rotations += 1
            current = current.next

        return rotations

def createList(arr):
    ll = LinkedList()
    for num in arr:
        ll.append(num)
    return ll.head

def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")
def rotation_times(head):
    ll = LinkedList()
    ll.head = head
    rotations = ll.find_right_rotations()
    return rotations

print('==============Test Case 1=============')
head = createList(np.array([13, 10, 6, 20, 17]))
print('Rearranged Necklace: ', end=' ')
printLinkedList(head)
returned_value = rotation_times(head)
print(f'Rotated: {returned_value} times')  # This should print 3
unittest.output_test(returned_value, 3)
print()

```

```python
import numpy as np

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

def createList(arr):
    linked_list = LinkedList()
    for item in arr:
        linked_list.append(item)
    return linked_list

def printLinkedList(head):
    current = head.head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")

def remove_compartment(head, n):
    current = head.head
    last_occurrence = None
    while current:
        if current.data == n:
            last_occurrence = current
        current = current.next
    if last_occurrence:
        current = head.head
        if current == last_occurrence:
            head.head = last_occurrence.next
        else:
            while current.next != last_occurrence:
                current = current.next
            current.next = last_occurrence.next
    return head

print('==============Test Case 1=============')
head = createList(np.array([14,10,15,10,41,10,72]))
number = 10
print('Original Compartment Sequence: ', end = ' ')
printLinkedList(head)
head = remove_compartment(head, number)
print('Changed Compartment Sequence: ', end = ' ')
printLinkedList(head)
print()

print('==============Test Case 2=============')
head = createList(np.array([10,15,33,41,14,72]))
number = 10
print('Original Compartment Sequence: ', end = ' ')
printLinkedList(head)
head = remove_compartment(head, number)
print('Changed Compartment Sequence: ', end = ' ')
printLinkedList(head)
print()

print('==============Test Case 3=============')
head = createList(np.array([10,15,33,41,14,72]))
number = 56
print('Original Compartment Sequence: ', end = ' ')
printLinkedList(head)
head = remove_compartment(head, number)
print('Changed Compartment Sequence: ', end = ' ')
printLinkedList(head)
print()
```

```python
import numpy as np

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

def createList(arr):
    linked_list = LinkedList()
    for item in arr:
        linked_list.append(item)
    return linked_list

def printLinkedList(head):
    current = head.head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")

def capture_the_flag(head):
    current = head.head
    position = 1
    flag_head = None
    flag_current = None

    while current is not None:
        if current.data % position == 0:
            if flag_head is None:
                flag_head = Node(current.data // position)
                flag_current = flag_head
            else:
                flag_current.next = Node(current.data // position)
                flag_current = flag_current.next
            position += 1
            current = current.next
        else:
            return "Linkwise"

    if flag_head is not None:
        return flag_head
    else:
        return "Linkwise"
print('==============Test Case 1=============')
head = createList(np.array([11, 8, 21, 20, 5, 42]))
print('Original Sequence: ', end=' ')
printLinkedList(head)
new_head = capture_the_flag(head)
print('The flag is: ', end=' ')
if isinstance(new_head, str):
    print(new_head)
else:
    current = new_head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")
print()

print('==============Test Case 2=============')
head = createList(np.array([11, 8, 28, 20, 5, 42]))
print('Original Sequence: ', end=' ')
printLinkedList(head)
new_head = capture_the_flag(head)
print('The flag is: ', end=' ')
if isinstance(new_head, str):
    print(new_head)
else:
    current = new_head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")
print()
```

Shuffle on song

```python
import numpy as np
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
def shuffle_on_song(head):
    even_head = LinkedList()
    odd_head = LinkedList()
    current = head.head
    while current:
        if ord(current.data) % 2 == 0:
            even_head.append(current.data)
        else:
            odd_head.append(current.data)
        current = current.next
    if even_head.head:
        current = even_head.head
        while current.next:
            current = current.next
        current.next = odd_head.head
        result = LinkedList()
        result.head = even_head.head
        return result
    else:
        return odd_head
def print_linked_list(linked_list):
    current = linked_list.head
    while current:
        print(current.data, end="→")
        current = current.next
    print("None")
def createList(arr):
    ll = LinkedList()
    for item in arr:
        ll.append(item)
    return ll

print('==============Test Case 1=============')
head = createList(np.array(['S','E','N','P','A','I']))
print('Original Play List: ', end=' ')
print_linked_list(head)
newhead = shuffle_on_song(head)
print('New Play List: ', end=' ')
print_linked_list(newhead)  # This should print N-->P-->S-->E-->A-->I
print()

print('==============Test Case 2=============')
head = createList(np.array(['N','I','S','H','I','N','O','Y','A']))
print('Original Play List: ', end=' ')
print_linked_list(head)
newhead = shuffle_on_song(head)
print('New Play List: ', end=' ')
print_linked_list(newhead)  # This should print N-->H-->N-->I-->S-->I-->O-->Y-->A
print()
```
