```python
# TASK 1
class Patient:
  def __init__(self, id1 = None, name = None, age = None, bloodgroup = None, prev = None, next = None):
    self.id1 = id1
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.prev = prev
    self.next = next

class WRM:
  def __init__(self):
    self.head = Patient()
    self.head.prev = self.head
    self.head.next = self.head

  def RegisterPatient(self, id1 = None, name = None, age = None, bloodgroup = None):
    new_patient = Patient(id1, name, age, bloodgroup)
    new_patient.next = self.head
    new_patient.prev = self.head.next
    self.head.prev.next = new_patient
    self.head.prev = new_patient
    print("PATIENT SUCCESSFULLY REGISTERED!")

  def ServePatient(self):
    if self.head.next != self.head:
      patient_served = self.head.next
      print("Patient being served:", patient_served.name)
      self.head.next = patient_served.next
    else:
      print("NO PATIENTS TO BE SERVED.")

  def CancelAll(self):
    self.head.next = self.head
    self.head.prev = self.head
    print("ALL APPOINTMENTS ARE CANCELLED.")

  def CanDoctorGoHome(self):
    if self.head.next == self.head:
      print("YES, there are no patients in the waiting room.")
    else:
      print("NO, there are patients waiting to be served.")

  def ShowAllPatient(self):
    if self.head.next != self.head:
      temp = self.head.next
      while temp.next != self.head:
        print(temp.id1, end = ", ")
        temp = temp.next
      while temp != self.head:
        print(temp.id1, end = " ")
        temp = temp.next
      print()
    else:
      print("THERE ARE NO PATIENTS IN THE LIST.")

tester_class = WRM()
while True:
  print("="*26)
  print("WELCOME TO WAITING ROOM MANAGEMENT. PLEASE ENTER YOUR CHOICE FROM THE FOLLOWING:")
  print("To register a patient, enter 1")
  print("To serve a patient, enter 2")
  print("To cancel all the appointments, enter 3")
  print("To know whether the doctor can go for lunch, enter 4")
  print("To get the list of patients' IDs in the queue, enter 5")
  print("To exit, enter 6")
  print("="*26)
  choice = int(input("Enter your choice: "))

  if choice == 1:
    print("PLEASE ENTER THE FOLLOWING DETAILS TO REGISTER THE PATIENT: ")
    id1 = int(input("Enter Patient's ID: "))
    name = input("Enter Patient's Name: ")
    age = int(input("Enter Patient's Age: "))
    bloodgroup = input("Enter Patient's Blood Group: ")
    print("PLEASE WAIT WHILE THE PATIENT IS BEING REGISTERED...")
    tester_class.RegisterPatient(id1, name, age, bloodgroup)
  elif choice == 2:
    tester_class.ServePatient()
  elif choice == 3:
    tester_class.CancelAll()
  elif choice == 4:
    tester_class.CanDoctorGoHome()
  elif choice == 5:
    tester_class.ShowAllPatient()
  elif choice == 6:
    print("THANKS FOR USING OUR SERVICE.")
    break
```
