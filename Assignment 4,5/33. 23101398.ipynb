```python
class Patient:
    def __init__(self, id=None, name=None, age=None, bloodgroup=None, next=None, prev=None):
        self.id, self.name, self.age, self.bloodgroup, self.next, self.prev = id, name, age, bloodgroup, next, prev

class WRM:
    def __init__(self):
        self.head = Patient()
        self.head.next = self.head
        self.head.prev = self.head

    def RegisterPatient(self, id, name, age, bloodgroup):
        last_patient = self.head.prev
        new_patient = Patient(id, name, age, bloodgroup, self.head, last_patient)
        self.head.prev = new_patient
        last_patient.next = new_patient
        print("Successfully added a new patient.")

    def ServePatient(self):
        if self.head.next == self.head:
            print("No patients to serve.")
        else:
            served = self.head.next
            next_patient = served.next
            self.head.next = next_patient
            next_patient.prev = self.head
            print(f"{served.name} got served.")

    def CancelAll(self):
        if self.head.next == self.head:
            print("No patients remaining to cancel.")
        else:
            self.head.next = self.head
            self.head.prev = self.head
            print("All appointments canceled. Doctor can go to lunch.")

    def CanDoctorGoHome(self):
        if self.head.next == self.head:
            return "yes"
        else:
            return "no"

    def ShowAllPatient(self):
        temp = self.head.next
        while temp != self.head:
            print(f"{temp.name} patient")
            temp = temp.next

    def ReverseTheLine(self):
        if self.head.next == self.head:
          print("No patients to reverse.")
          return

        cur = self.head.prev
        while cur != self.head:
          print(cur.name)
          cur = cur.prev
wrm = WRM()

while True:
    print("Welcome to Waiting Room Management - Choose an Option:")
    print("1. RegisterPatient")
    print("2. ServePatient")
    print("3. CancelAll")
    print("4. CanDoctorGoHome")
    print("5. ShowAllPatient")
    print("6. ReverseTheLine")
    print("7. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter ID: "))
        name = input("Enter Name: ")
        age = int(input("Enter Age: "))
        blood_group = input("Enter Blood Group: ")
        wrm.RegisterPatient(id, name, age, blood_group)
        print("Executing RegisterPatient...")

    elif choice == 2:
        wrm.ServePatient()
        print("Executing ServePatient...")

    elif choice == 3:
        wrm.CancelAll()
        print("Executing CancelAll...")

    elif choice == 4:
        result = wrm.CanDoctorGoHome()
        if result == "yes":
            print("Yes, the doctor can go home.")
        else:
            print("No, patients are waiting.")
        print("Executing CanDoctorGoHome...")

    elif choice == 5:
        print("All patients in sequence to serve:")
        wrm.ShowAllPatient()
        print("Executing ShowAllPatient...")

    elif choice == 6:

        wrm.ReverseTheLine()
        print("Executing ReverseTheLine...")

    elif choice == 7:
        print("Exiting the program.")
        break
```
