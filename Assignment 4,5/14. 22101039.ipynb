```python
class Patient:
  def __init__(self,id,name,age,btype,next=None,prev=None):
    self.id=id
    self.name=name
    self.age=age
    self.btype=btype
    self.next=next
    self.prev=prev

class WRM:
  def __init__(self):
    self.head=Patient(id=None,name=None,age=None,btype=None,next=None,prev=None) #dummy head
    # self.tail=self.head
    self.head.prev=self.head
    self.head.next=self.head

  def registerPatient(self,id,name,age,btype):
    # print(" Paitent resistering  ")
    last=self.head.prev #dh address
    new=Patient(id,name,age,btype)
    new.next=self.head
    self.head.prev=new
    new.prev=last
    last.next=new
    # print(" Paitent resistering  ")

  def servePatient(self):
    if self.head==self.head.next:
      print("No patient to be served")
      return
    else:
      patient=self.head.next
      self.head.next=patient.next
      patient.next.prev=self.head
      print(f"{patient.name} is served")

  def showAllPatient(self):
    if self.head==self.head.next:
      print("No patient in the WRM")
      return
    else:
      temp=self.head.next
      while temp!=self.head:
        print("Name:",temp.name)
        temp=temp.next

  def canDoctorGoHome(self):
    if self.head==self.head.next:
      return True
      # print("No patients in the waiting room.")
    else:
      return  False
      # print("no")

  def cancelAll(self):
    if self.head==self.head.next:
      print("No patient")
    else:
      temp=self.head.next
      temp2=self.head.prev
      self.head.next=self.head
      self.head.prev=self.head
      temp.prev=None
      temp2.next=None
      print("all  appointments are canceled")


  def ReverseTheLine(self):
    if self.head==self.head.next:
      print("No patient in WRM")
      return
    temp=self.head.prev #last node
    oldprev=temp.next
    oldnext=temp.prev # previous node
    while temp != self.head:

      temp.prev,temp.next=temp.next,temp.prev
      if temp ==self.head:
        break

    self.head,self.head.next=self.head.next,self.head
  # def ReverseTheLine(self):
  #   if self.head == self.head.next:
  #       print("No patients in the WRM")
  #       return

  #   current = self.head
  #   while True:
  #       current.next, current.prev = current.prev, current.next
  #       current = current.prev
  #       if current == self.head:
  #           break

  #   self.head, self.head.next = self.head.next, self.head


def displayOptions():
  print("\n----- Waiting Room Management System Menu -----")
  print("1. Add Patient")
  print("2. Serve Patient")
  print("3. Show All Patients")
  print("4. Can Doctor Go Home?")
  print("5. Cancel All Appointments")
  print("6. ReverseTheLine")
  print("7. Exit")

sample=WRM()
while True:
  displayOptions()
  user=int(input("enter "))
  if user == 1:
    id = int(input("Enter patient ID: "))
    name = input("Enter patient name: ")
    age = int(input("Enter patient age: "))
    btype= input("Enter patient blood group: ")
    sample.registerPatient(id, name, age, btype)
    print("Patient added successfully.")

  elif user == 2:
    sample.servePatient()
  elif user == 3:
    sample.showAllPatient()
  elif user == 4:
    if sample.canDoctorGoHome():
      print("Yes, doctor can go home.")
    else:
      print("No, patients are waiting.")
  elif user ==5:
    sample.cancelAll()
  elif user == 6:
    sample.ReverseTheLine()
  elif user == 7:
    print("Exiting...")
    break
```
