```python
class Patient:
    def __init__(self, patient_id, name, age, blood_group, next_patient, prev_patient):
        self.name = name
        self.patient_id = patient_id
        self.age = age
        self.blood_group = blood_group
        self.next = next_patient
        self.prev = prev_patient

class WRM:
    def __init__(self):
        self.dh = Patient(None, None, None, None, None, None)
        self.dh.prev = self.dh
        self.dh.next = self.dh
        self.tail = self.dh

    def RegisterPatient(self, patient_id, name, age, blood_group, room):
        patient = Patient(patient_id, name, age, blood_group, self.dh, self.tail)
        self.tail.next = patient
        self.tail = self.tail.next
        self.dh.prev = patient
        self.room = 'Waiting room ' + room
        print(f'Successfully added patient to {self.room}\nName:{patient.name},  Id:{patient.patient_id}\nAge:{patient.age},  Blood group:{patient.blood_group}\n============================')

    def ShowAllPatient(self):
        temp = self.dh.next
        if temp == self.dh:
            print(f'There are no patients in this room\n============================')
            return None
        else:
            print(f'Patients Waiting to be served in {self.room}: ')
            while temp != self.dh:
                print(f'Name:{temp.name},  Id:{temp.patient_id}\nAge:{temp.age},  Blood group:{temp.blood_group}\n============================')
                temp = temp.next

    def ServePatient(self):
        if self.dh.next == self.dh:
            print(f'No patients to serve in this room\n============================')
        else:
            patient_to_serve = self.dh.next
            print(f'{patient_to_serve.name} is being served\n============================')
            next_patient = patient_to_serve.next
            patient_to_serve.next = None
            patient_to_serve.prev = None
            self.dh.next = next_patient
            next_patient.prev = self.dh

    def CancelAll(self):
        if self.dh.next == self.dh:
            print(f'There are no appointments to cancel\n============================')
        else:
            print(f'All appointments in {self.room} have been cancelled\n============================')
        self.dh.next = self.dh
        self.dh.prev = self.dh

    def CanDoctorGoHome(self):
        temp = self.dh.next
        if temp == self.dh:
            return True
        else:
            return False

    def ReverseTheLine(self):
        if self.dh.next == self.dh:
            print("There is no patient line to reverse\n============================")
        else:
            print('Reversed patient line:')
            temp = self.dh.prev
            while temp != self.dh:
                print(f'Name:{temp.name},  Id:{temp.patient_id}\nAge:{temp.age},  Blood group:{temp.blood_group}\n============================')
                temp = temp.prev


room1=WRM()
room1.ShowAllPatient()
room1.CancelAll()
room1.ServePatient()
print(room1.CanDoctorGoHome())
print('============================')
room1.RegisterPatient('001','Mahir','21','A-','1')
room1.RegisterPatient('002','Abeg','22','O+','1')
room1.RegisterPatient('003','Orpa','23','AB+','1')
room1.RegisterPatient('004','Ashika','22','B+','1')
room1.RegisterPatient('005','Hasin','23','A+','1')
room2=WRM()
room2.RegisterPatient('001','Mim','20','O-','2')
room2.RegisterPatient('002','Dion','18','O-','2')
room2.RegisterPatient('003','Fahim','18','A+','2')
room2.RegisterPatient('004','Borsha','17','A-','2')
room2.RegisterPatient('005','Neha','21','AB+','2')
room2.ServePatient()
room2.CancelAll()
room2.ShowAllPatient()
print(room2.CanDoctorGoHome())
print('============================')
room1.ShowAllPatient()
print(room1.CanDoctorGoHome())
print('============================')
room1.ReverseTheLine()
```
