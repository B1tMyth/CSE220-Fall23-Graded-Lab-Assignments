```python
class Patient:
  def __init__(self,id=None,name=None,age=None,bld=None,next=None,prev=None):
    self.id,self.name,self.age,self.bld,self.next,self.prev=id,name,age,bld,next,prev
class WRM:
  def __init__(self):
    self.head=Patient()
    self.head.next=self.head
    self.head.prev=self.head
    self.tail=self.head
  def RegisterPatient(self,id,name,age,bld):
    new_patient=Patient(id,name,age,bld,None,None)
    new_patient.prev=self.tail
    self.tail.next=new_patient
    new_patient.next=self.head
    self.head.prev=new_patient
    self.tail=self.tail.next
  def ServePatient(self):

    if self.head.next==self.head:
      print("No patient is availble")
    else:
      served=self.head.next
      next_patient=served.next
      self.head.next=next_patient
      next_patient.prev=self.head
      print(served.name)
  def CancelAll(self):
    if self.head.next==self.head:
      print("already cancelled")
    else:
      self.head.next=self.head
      self.head.prev=self.head
  def CanDoctorGoHome(self):
    if self.head.next==self.head:
      return("Yes")
    else:
      return ("False")
  def ShowAllPatient(self):
    if self.head.next==self.head:
      return ("there is no patient")
    else:
      temp=self.head.next
      while temp!=self.head:
        print(temp.name)
        temp=temp.next
  def ReverseTheLine(self):
    cur=self.head.next
    while cur.next!=self.head:
      next_node=cur.next
      cur.next=cur.prev
      cur.prev=next_node
      cur=next_node
    temp=self.head
    cur.prev=self.head
    cur.next=temp.next
    self.head.prev=temp
    self.head.next=cur

ckd=WRM()

menu=(f'1.Register patient \n2. Serve patient \n3 Cancel all \n4. Can doctor go home \n5. Show all patients \n6. Stop')
while True:
  print("----------------------------------------------------------------------")
  print("press 1 for registering patient")
  print("press 2 for to know who are being served")
  print("press 3 for cancell all")
  print("press 4 to cehck the doc can go home or not")
  print("press 5 to show all patient")
  print("press 6 to reverse the line")
  print("press 7 to stop")
  pointer= ' '
  pointer= int(input("plese entr the option"))
  if pointer == 1:
    name=input("pls enter the name: ")
    id=input("pls enter id: ")
    age=input("pls enter the age: ")
    bld=input("pls enter the bloodgroup: ")
    if name==" " or id==" " or age==" " or bld==" ":
      print("not sucess")
    else:
      ckd.RegisterPatient(int(id),name,int(age),bld)
  elif pointer == 2:
    ckd.ServePatient()
  elif pointer ==3:
    ckd.CancelAll()
  elif pointer ==4:
    print(ckd.CanDoctorGoHome())
  elif pointer ==5:
    ckd.ShowAllPatient()
  elif pointer == 6:
    ckd.ReverseTheLine()
  else:
    break
```
