```python
class Patient:
   def __init__(self, id = None, name = None, age = None, blood_group = None, next = None, previous = None):
      self.id = id
      self.name = name
      self.age = age
      self.blood_group = blood_group
      self.next = next
      self.previous = previous
class WRM:
    def __init__(self):
      self.head = Patient()
      self.head.previous = self.head
      self.head.next = self.head
      self.size = 0
    def RegisterPatient(self,id,name=None,age=None,blood_group=None):
      patient_info=Patient(id,name,age,blood_group)
      patient_info.next = self.head
      patient_info.previous = self.head.next
      self.head.previous.next = patient_info
      self.head.previous = patient_info
      self.size +=1
    def ServePatient(self):
      if self.head.next != self.head:
          serve_patient = self.head.next
          print(f'Patient being served : {serve_patient.id} : {serve_patient.name}')
          self.head.next = serve_patient.next
          self.head.next.previous = serve_patient.previous
      elif self.head.next == self.head:
         print(f'No Patient to serve')
    def ShowAllPatient(self):
      if self.head.next != self.head:
          now = self.head.next
          while now !=self.head:
              print(now.name)
              now = now.next
          print()
    def CancelAll(self):
      self.head.next = self.head.previous = self.head
      print('All appointments canceled')
    def CanDoctorGoHome(self):
      if self.head.next==None:
        return f"Yes,he can go home"
      else:
        return f"NO, it is currenty not possible"
    def ReverseTheLine(self):
      if self.head.next != self.head:
        current = self.head.previous
        while current != self.head:
          print(current.name,end=" ")
          current = current.previous




# Driver code
tester = WRM()

while True:
    print()
    print("Let's proceed ... ... ...")
    print('1. Add Patient')
    print('2. Serve Patient')
    print('3. Show All Patients')
    print('4. Can Doctor Go Home')
    print('5. Cancel All Appointments')
    print('6. Reverse the Appointment list')
    print('7. Exit')
    print('====================================')
    print()
    option = int(input("Please select any option: "))
    print()
    if option == 1:
      print('Executing Registration......')
      id = input('Please enter Patient_ID : ')
      name = input('Please enter Name : ')
      age = input('Please enter AGE : ')
      blood_group = input('Please enter Bloodgroup : ')
      tester.RegisterPatient(id,name,age,blood_group)
      print('Registration Completed.....')
    elif option == 2:
      print(f'Serving patient ... ... ...')
      print()
      tester.ServePatient()
    elif option == 3:
      print(f'Showing the patient list ... ... ...')
      print()
      tester.ShowAllPatient()
    elif option == 4:
      print(f'Checking if the doctor can go home or not ... ... ...')
      print()
      print(tester.CanDoctorGoHome())
    elif option == 5:
      print(f'Canceling all the Appointments ... ... ...')
      print()
      tester.CancelAll()
    elif option == 6:
      print(f'Reversing the order ... ... ...')
      print()
      tester.ReverseTheLine()
    elif option ==7:
      print("Thank you for using our service")
      break
```
