```python
class DoublyList:

  # Creates a Dummy Head for Dummy Headed Circular Doubly Circular Linked List.
  def __init__(self):
        self.head = Node(None, None, None)  # Initialize with a dummy node pointing to itself


  #create a DHDCL from an array
  def create_list(self, a):
        self.head.next = self.head
        self.head.prev = self.head
        tail = self.head

        for i in range(len(a)):
            n = Node(a[i], self.head, tail)
            tail.next = n
            tail = n
            self.head.prev = tail

        return self.head

  # Counts the number of Nodes in the list and return the number
  def countNode(self):
        counter = 0
        h_next = self.head.next
        while h_next != self.head:
            counter += 1
            h_next = h_next.next

        return counter

    # To Do
    # Remove this line


  # prints the elements in the list forward
  def forwardprint(self):
    h_next = self.head.next
    while (h_next != self.head):
        if h_next is not None:
            if h_next is self.head.prev:
                print(h_next.element)
            else:
                print(h_next.element, end=', ')
        h_next = h_next.next


  # prints the elements in the list backward
  def backwardprint(self):
        h_prev = self.head.prev
        while (h_prev != self.head):
            if h_prev is self.head.next:
                print(h_prev.element)
            else:
                print(h_prev.element, end=', ')
            h_prev = h_prev.prev



  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    if idx < 0 or idx >= self.countNode():
        print("Index error")
        return None

    h_next = self.head.next  # Skip the dummy head
    finder = 0

    while h_next != self.head:
        if finder == idx:
            return h_next
        finder += 1
        h_next = h_next.next

    return None  # Handle any other cases



  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
        h_next = self.head.next
        counter = 0
        while h_next != self.head:
            if h_next.element == elem:
                return counter
            else:
                counter += 1
                h_next = h_next.next
        return -1


  # insert at specific index of the DHDCL
  def insert(self, elem, idx):
        if idx < 0 or idx > self.countNode():
            return None
        else:
            node = Node(elem, None, None)
            h_next = self.head.next
            counter = 0
            while counter < idx:
                h_next = h_next.next
                counter = counter + 1
            p = h_next.prev
            node.next = h_next
            h_next.prev = node
            p.next = node
            node.prev = p


  #removes the node of the specific index and returns the element of the node
  def remove(self, idx):
        if idx < 0 or idx > self.countNode() - 1:
            return None
        else:
            h_counter = self.head.next
            counter = 0
            while counter < idx:
                h_counter = h_counter.next
                counter += 1
            remv = str(h_counter.element)
            p = h_counter.prev
            n = h_counter.next
            p.next = n
            n.prev = p
            h_counter = None

        return remv
```
