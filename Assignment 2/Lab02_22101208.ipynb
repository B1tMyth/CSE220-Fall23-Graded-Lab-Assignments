```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
    for i in range(len(floor[0])):
        if i % 2 == 0:
            for j in range(0, len(floor), 2):
                print(floor[j][i], end = " ")

        else:
            if len(floor) % 2 == 0:
                for j in range(len(floor)-1, 0, -2):
                    print(floor[j][i], end = " ")

            else:
                for j in range(len(floor)-2, 0, -2):
                    print(floor[j][i], end = " ")

        print()


floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

```python
def create_fence(district, depth):
    district_row, district_col = district.shape
    arr1 = np.full(((district_row + 2*depth) , (district_col)+2*depth),8)
    for i in range(district_row):
        for j in range(district_col):
            arr1[i+depth][j+depth]=district[i][j]
    arr2=np.zeros
    return arr1

depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
  row, col = clubroom.shape
  length = (row * (row - 1)) // 2
  strength_diff = np.array([0]*length)

  idx = 0
  for i in range(row):
    for j in range(i+1, col):
            diff = clubroom[i][j] - clubroom[j][i]
            strength_diff[idx] = diff
            idx += 1
  return strength_diff

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
    arr = np.zeros((n, n), dtype=int)
    for i in range(n):
        for j in range(i+1):
            if j == i or j == 0:
                arr[i][j] = i + 1

            else:
                sum = 0
                for k in range(j+1):
                    sum += arr[i - 1][k]

                arr[i][j] = sum

    return arr

def print_without_none(matrix):
    for i in range(len(matrix)):
        for j in range(i+1):
            print(matrix[i][j], end = " ")

        print()



n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```
