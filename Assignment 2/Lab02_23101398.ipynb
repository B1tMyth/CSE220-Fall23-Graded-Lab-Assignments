#Wall Up Trost District

```python
import numpy as np
def print_matrix(matrix):
  for row in matrix:
    print(' '.join([str(x) for x in row]))
def create_fence(district, depth):
    district_row, district_col = district.shape
    new_rows = district_row + 2 * depth
    new_cols = district_col + 2 * depth


    fenced_district = np.array([[0] * new_cols] * new_rows)

    for i in range(district_row):
      for j in range(district_col):

        if i < depth or i > new_rows - depth or j < depth or j >= new_cols - depth:
         fenced_district[i][j] = 0
        else:

         fenced_district[i + depth][j + depth] = district[i][j]

    return fenced_district

depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
import numpy as np
def print_matrix(matrix):

  for row in matrix:
    print(' '.join([str(x) for x in row]))

def calculate_strength_diff(clubroom):
    rows, cols = len(clubroom), len(clubroom[0])
    size = rows * (rows + 1) // 2
    strength_diff = [0] * size

    idx = 0
    for i in range(rows):
        for j in range(cols):
            if i < j:
                strength_diff[idx] = clubroom[i][j] - clubroom[j][i]
                idx += 1
    return strength_diff

def strength_difference(clubroom):
    strength_diff = calculate_strength_diff(clubroom)
    return strength_diff


clubroom = np.array([
    [1, 2, 9, 7],
    [4, 5, 1, 8],
    [3, 6, 2, 7],
    [2, 8, 6, 3]
])

print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is:', returned_value)
```

#ATM's Triangle

```python
def atm_triangle(n):
    atm_triangle = []

    for i in range(1, n + 1):
        row = [0] * i
        row[0] = row[-1] = i

        if i > 2:
            for j in range(1, i - 1):
                row[j] = atm_triangle[i - 2][j - 1] + atm_triangle[i - 2][j]

        atm_triangle.append(row)
    return atm_triangle

def print_without_none(matrix):
    for row in matrix:
        for element in row:
            if element is not None:
                print(element, end=" ")
        print()
n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
```

#Trace The BOT

```python
import numpy as np

def moving_around(cmds):
    grid = np.full((7, 7), '.', dtype=str)
    row, col = 3, 3

    grid[row, col] = '/'

    for cmd in cmds:
        if cmd == 1 and row > 1 and col > 1:
            row -= 2
            col -= 1
        elif cmd == 2 and row > 1 and col < 5:
            row -= 2
            col += 1
        elif cmd == 3 and row > 2:
            row -= 2
        elif cmd == 4 and row < 4:
            row += 2
        elif cmd == 5 and row > 1 and col > 1:
            row -= 2
            col -= 1
        elif cmd == 6 and row > 1 and col < 5:
            row -= 2
            col += 1
        elif cmd == 7 and row > 2:
            row -= 2
        elif cmd == 8 and row < 4:
            row += 2
        elif cmd == 9 and row < 5 and col > 1:
            row += 2
            col -= 1
        elif cmd == 10 and row < 5 and col < 5:
            row += 2
            col += 1
        elif cmd == 11 and row < 6:
            row += 2
        elif cmd == 12 and row < 6:
            row += 2
        grid[row, col] = '*'

    return grid
def print_matrix(matrix):
    for row in matrix:
        print(" ".join(row))
    print("-" * 45)
cmds = np.array([5, 11, 2, 9])
result = moving_around(cmds)
print_matrix(result)
```
