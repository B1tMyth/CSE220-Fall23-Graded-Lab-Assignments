```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  #To Do
  row,col = floor.shape
  idx = 0

  for i in range(col):
    if i % 2 == 0:
      for j in range(idx,row,2):
        print(floor[j][i], end=' ')
        idx = row - 2
      print()
    else:
      if row % 2 != 0:
        for j in range(idx,-1,-2):
          print(floor[j][i], end=' ')
          idx = 0
        print()

      else:
        idx = row - 1
        for j in range(idx, 0 , -2):
          print(floor[j][i], end=' ')
        print()
        idx = 0



floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    district_row, district_col = district.shape
    row = district_row + 2 * depth
    col = district_col + 2 * depth
    fence = np.full((row,col),8)
    for i in range(len(district)):
      for j in range( len(district[i]) ):
        fence[i + depth][j + depth] = district[i][j]

    return fence

depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
  strength_diff = np.zeros(100, dtype = int)
  row , col = clubroom.shape
  strength_diff = np.array([0]*int((((row*col)-4)/2)))
  idx = 0

  for i in range(row):
    for j in range(i,col):
      if i!=j:
        strength_diff[idx]=clubroom[i][j]-clubroom[j][i]
        idx += 1
  return strength_diff

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
  #pass
  matrix =np.full((n,n),None)
  x=-2
  for i in range(1,n+1):
    if i==1:
      matrix[x+2][0]=x+3
      x+=1
    elif i==2:
      matrix[x+1][0]=2
      matrix[x+1][x+1]=2
      x+=1
    else:
      for j in range(1,i-1):
        matrix[x][0]=x+1
        matrix[x][x]=x+1
        y=0
        for k in range(0,j+1):
          y += matrix[x-1][k]
        matrix[x][j] = y
    x+=1
  return matrix

def print_without_none(matrix):
  out=''
  for i in range(n):
    y=''
    for j in range(n):
      if matrix[i][j]!=None:
        y+=str(matrix[i][j])+'  '
    out += f'\n{y}'
  print(out)



n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```
