```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  row,col=floor.shape
  for j in range(col):
    if j%2==0:
      for i in range(0,row,2):
        print(floor[i][j],end=" ")
    else:
      if row%2!=0:
        for i in range(row-2,-1,-2):
          print(floor[i][j],end=" ")
      else:
        for i in range(row-1,-1,-2):
          print(floor[i][j],end=" ")
    print()


floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    district_row, district_col = district.shape
    fence_row = district_row + 2 * depth
    fence_col = district_col + 2 * depth
    fenced_district=np.zeros((fence_row,fence_col),dtype=int)
    for a in range(fence_row):
      for k in range(fence_col):
        fenced_district[a][k]=8

    for i in range(district_row):
        for j in range(district_col):
            fenced_district[i + depth][j + depth] = district[i][j]

    return fenced_district

depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
  row,col=clubroom.shape
  length=(row**2-row)//2
  strength_diff = np.zeros(length, dtype = int)
  index=0
  for i in range(row):
    for j in range(col):
      if i<=j and i!=j:
        strength_diff[index]=clubroom[i][j]-clubroom[j][i]
        index+=1

  return strength_diff

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
  arr = np.zeros((n,n), dtype=int)
  for i in range(n):
    arr[i][0] = i+1
    arr[i][i] = i+1
  for i in range(2,n):
    for j in range(1,i):
      arr[i][j] = arr[i-1][j-1] + arr[i-1][j]
  return arr

def print_without_none(matrix):
  row,col=matrix.shape
  for a in range(row):
    for b in range(col):
      if matrix[a][b]!=0:
        print(matrix[a][b],end=" ")
    print()

n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

```python
def moving_around(cmds):
  grid = np.full((7, 7), '.')
  row=3
  column=3
  grid[row][column]="-"

  for i in range(len(cmds)):
    if cmds[i]==1:
      if row-2>=0 and  column-3>=0:
          grid[row-2][column-3]="*"
          row-=2
          column-=3
    elif cmds[i]==2:
      if row-1>=0 and  column-2>=0:
          grid[row-2][column-1]="*"
          row-=2
          column-=1
    elif cmds[i]==3:
      if row-3>=0 and  column-2>=0:
          grid[row-3][column-2]="*"
          row-=3
          column-=2
    elif cmds[i]==4:
      if row-1>=0 and  column-2>=0:
          grid[row-1][column-2]="*"
          row-=1
          column-=2
    elif cmds[i]==5:
      if row-2>=0 and column+1<=6:
        grid[row-2][column+1]="*"
        row-=2
        column+=1

    elif cmds[i]==6:
      if row-2>=0 and column+3<=6:
        grid[row-2][column+3]="*"
        row-=2
        column+=3
    elif cmds[i]==7:
      if row-3>=0 and column+2<=6:
        grid[row-3][column+2]="*"
        row-=3
        column+=2
    elif cmds[i]==8:
      if row-1>=0 and column+2<=6:
        grid[row-1][column+2]="*"
        row-=1
        column+=2
    elif cmds[i]==9:
      if row+2<=6 and column-3>=0:
        grid[row+2][column-3]="*"
        row+=2
        column-=3
    elif cmds[i]==10:
      if row+2<=6 and column-1>=0:
        grid[row+2][column-1]="*"
        row+=2
        column-=1
    elif cmds[i]==11:
      if row+1<=6 and column-2>=0:
        grid[row+1][column-2]="*"
        row+=1
        column-=2
    elif cmds[i]==12:
      if row+3<=6 and column-2>=0:
        grid[row+3][column-2]="*"
        row+=3
        column-=2
  grid[row][column]="/"


  return grid

cmds = np.array([5,11,2,9])
result = moving_around(cmds)
print_matrix(result)
#This should print
# -------------------------------------------
# |  .  |  /  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  *  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  *  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  -  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
```
