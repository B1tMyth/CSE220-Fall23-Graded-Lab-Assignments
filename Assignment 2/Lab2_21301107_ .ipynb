```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  row,col=floor.shape
  for i in range(0,col,1):
    if i%2==0:
      for j in range(0,row,2):
        print(floor[j][i])
    else:
      for j in range(len(floor)-2,-1,-2):
        print(floor[j][i])




floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Crows vs Cats

```python
def strength_difference(clubroom):
  strength_diff = np.zeros(100, dtype = int)
  #To Do
  s=clubroom.shape[0]
  size_strength_diff=(s*(s-1))//2
  strength_diff=np.zeros(size_strength_diff,dtype=int)
  i=0
  for i in range(s):
    for j in range(i+1,s):
      strength_diff[i]=clubroom[i,j]-clubroom[j,i]
      i+=1


  return strength_diff

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```
