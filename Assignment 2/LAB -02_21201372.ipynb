```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Crows vs Cats

```python
def strength_difference(clubroom):
    strength_diff = np.zeros(6, dtype=int)
    n = clubroom.shape[0]
    index = 0
    for i in range(n):
      for j in range(i + 1, n):
        strength_diff[index] = clubroom[i, j] - clubroom[j, i]
        index += 1

    return strength_diff

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
  triangle = np.zeros((n, n), dtype=int)
  for i in range(n):
    for j in range(i + 1):
      if i==j   :
        triangle[i][j]=i+1
        triangle[i][0]=i+1
      else:
        x=0
        for k in range (0,j+1):
          x+=triangle[i-1][k]
        triangle[i][j]=x
  return triangle
def print_without_none(matrix):
  for i in range(matrix.shape[0]):
    for j in range(i + 1):
      print(matrix[i, j], end=" ")
    print()


n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```
