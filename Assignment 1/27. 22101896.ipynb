```python
# Test 01: Play Right
def playRight(sequence, beats):
  k=0
  for i in beats:
    if i==1:
      k=k+1

  for i in range(k):
    temp=sequence[len(sequence)-1]
    for j in range(len(sequence)-2,-1,-1):
      sequence[j+1]=sequence[j]
    sequence[0]=temp
  return sequence



print("///  Test 01: Play Right  ///")
sequence = np.array([10, 20, 30, 40, 50, 60])
beats = np.array([1, 0, 0, 1, 0, 1])
returned_value = playRight(sequence, beats)
print(f'Task 1: {returned_value}')  # This should print [40, 50, 60, 10, 20, 30]
```

```python
 #Test 03: Merge Lineup

import numpy as np


def mergeLineup(pokemon_1, pokemon_2):
  # pokemon_1=np.copy(pokemon_1)
  # pokemon_2=np.copy(pokemon_2)

  result = [None]*len(pokemon_1) #this array stores the resulting array
  for i in range(len(pokemon_1)):
    if pokemon_1[i]==None:
      pokemon_1[i]=0
    if pokemon_2[i]==None:
      pokemon_2[i]=0

  for i in range(len(pokemon_1)):
    result[i]=pokemon_1[i]+pokemon_2[len(pokemon_1)-1-i]

  result=np.copy(result)
  return result

print("///  Test 03: Merge Lineup  ///")
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None] )
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 3: {returned_value}') # This should print [12, 3, 28, -8, 5]
unittest.output_test(returned_value, np.array([12, 3, 28, -8, 5]))

pokemon_1 = np.array([4, 5, -1, None, None])
pokemon_2 = np.array([2, 27, 7, 12, None])
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 3: {returned_value}') # This should print [4,17,6,27,2]
unittest.output_test(returned_value, np.array([4,17,6,27,2]))
```

```python
# Test 04: Balance your Salami
def balanceSalami(salami):
  k=0
  for j in salami:
      k=k+1

  lis1=0
  res1=0
  for a in range(0,k):
    res1 = res1 + salami[a]
  for m in range(0, k):
      lis1 = lis1 + salami[m]
      res1 = res1 - salami[m]
      if lis1 == res1:
        return True
  return False

print("///  Test 04: Balance Your Salami  ///")
salami = np.array([1, 1, 1, 2, 1])
returned_value = balanceSalami(salami)
print(f'Task 4: {returned_value}') # This should print True
unittest.output_test(returned_value, True)

salami = [2, 1, 1, 2, 1]
returned_value = np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}') # This should print False
unittest.output_test(returned_value, False)

salami = [10, 3, 1, 2, 10]
returned_value =  np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}') # This should print True
unittest.output_test(returned_value, True)
```

```python
#part2

import numpy as np


def get_mean(arr):
    return np.mean(arr)


def get_standard_deviation(arr):
    return np.std(arr, ddof=1)


def filter_values(arr, mean, std_dev):
    a = 1.5 * std_dev
    n_array = [x for x in arr if abs(x - mean) >= a]
    return n_array if len(n_array) > 0 else None


test_case = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])


mean_value = get_mean(test_case)
std_dev_value = get_standard_deviation(test_case)


filtered_values = filter_values(test_case, mean_value, std_dev_value)

print("Mean of the numbers is:", mean_value)
print("Standard deviation is:", std_dev_value)
print("New array:", filtered_values)
```
