```python
# Test 01: Play Right
def playRight(sequence, beats):
    n = len(sequence)


    for i in range(n):
        if beats[i] == 1:
            last_element = sequence[n-1]
            #print(sequence)
            #print(last_element)
            for j in range(n - 1, -1, -1):
                sequence[j] = sequence[j - 1]
            sequence[0] = last_element


    return sequence

print("///  Test 01: Play Right  ///")
sequence = np.array([10, 20, 30, 40, 50, 60])
beats = np.array([1, 0, 0, 1, 0, 1])
returned_value = playRight(sequence, beats)
print(f'Task 1: {returned_value}')  # This should print [40, 50, 60, 10, 20, 30]
```

```python
# Test 02: Discard Cards
def discardCards(cards,number):
  n = len(cards)
  count = 0

  for i in range(n):
        if cards[i] != number:
            cards[count] = cards[i]
            count += 1
  while count < n:
        cards[count] = 0
        count += 1
  return cards.tolist()
print("///  Test 02: Discard Cards  ///")
cards = np.array([1,2,3,2,8,2,2,5,7])
returned_value = discardCards(cards, 2)
print(f'Task 2: {returned_value}') # This should print [1, 3, 8, 5, 7, 0, 0, 0, 0]
unittest.output_test(returned_value, np.array([1, 3, 8, 5, 7, 0, 0, 0, 0]))
```

```python
# Test 04: Balance your Salami
def balanceSalami(salami):
  # TO DO
  total_salami = sum(salami)

  if total_salami % 2 != 0:
        return False

  half_of_salami = total_salami // 2
  current_total = 0

  for i in range(len(salami)):
        current_total += salami[i]
        if current_total == half_of_salami:
            return True
        if current_total > half_of_salami:
            return False
  return False


print("///  Test 04: Balance Your Salami  ///")
salami = np.array([1, 1, 1, 2, 1])
returned_value = balanceSalami(salami)
print(f'Task 4: {returned_value}') # This should print True
unittest.output_test(returned_value, True)

salami = [2, 1, 1, 2, 1]
returned_value = np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}') # This should print False
unittest.output_test(returned_value, False)

salami = [10, 3, 1, 2, 10]
returned_value =  np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}') # This should print True
unittest.output_test(returned_value, True)
```

```python
#Part2
import math
import numpy as np

def mean(arr):
    total = 0
    for x in arr:
        total += x
    return total / len(arr)

def std_deviation(arr):
    mean_value = mean(arr)
    squared_diff_sum = 0
    length = len(arr)

    for x in arr:
        squared_diff_sum += (x - mean_value) ** 2

    variance = squared_diff_sum / (length - 1)
    std_dev = math.sqrt(variance)

    return std_dev

def sd_away_1_5(arr):
    mean_value = mean(arr)
    std_dev = std_deviation(arr)
    count = 0
    for x in arr:
        sd_away = (x - mean_value) / std_dev
        if sd_away >= 1.5 or sd_away <= -1.5:
            count += 1
    if count == 0:
        return None

    new_arr = np.array([0] * count)
    idx = 0

    for x in arr:
        sd_away = (x - mean_value) / std_dev
        if sd_away >= 1.5 or sd_away <= -1.5:
            new_arr[idx] = x
            idx += 1
    return new_arr

numbers = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print("The mean of the numbers is: ", mean(numbers))
print("The standard deviation is: ", std_deviation(numbers))
print("New array: ", sd_away_1_5(numbers))
```
