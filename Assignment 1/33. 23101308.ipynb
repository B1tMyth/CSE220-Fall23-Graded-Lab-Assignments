```python
# Test 04: Balance your Salami
def balanceSalami(salami):
  total=0
  for i in salami:
    total+=i
  me=0
  sis=total
  done=False
  for j in range(len(salami)):
    me+=salami[j]
    sis-=salami[j]
    if me==sis:
      done=True

  return done

print("///  Test 04: Balance Your Salami  ///")
salami = np.array([1, 1, 1, 2, 1])
returned_value = balanceSalami(salami)
print(f'Task 4: {returned_value}') # This should print True
unittest.output_test(returned_value, True)

salami = [2, 1, 1, 2, 1]
returned_value = np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}') # This should print False
unittest.output_test(returned_value, False)

salami = [10, 3, 1, 2, 10]
returned_value =  np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}') # This should print True
unittest.output_test(returned_value, True)
```

```python
#part2
def mean(arr):
  total=0
  for i in arr:
    total+=i
  mean=total/len(arr)
  return mean

def deviation(arr):
  a=mean(arr)
  total=0
  for i in arr:
    x=(i-a)**2
    total+=x
  deviation=(total/(len(arr)-1))**0.5
  return deviation

def away(arr):
  x=np.array([None]*len(arr))
  up=mean(arr)+1.5*deviation(arr)
  low=mean(arr)-1.5*deviation(arr)
  count=0
  for i in arr:
    if i>up or i<low:
      x[count]=i
      count+=1
  new=np.array([None]*count)
  for i in range(len(new)):
    new[i]=x[i]
  return new

value=np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print("The mean of the numbers is:",mean(value))
print("The standard deviation is:",deviation(value).round(2))
print("New array:",away(value))
```
