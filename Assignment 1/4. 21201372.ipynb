```python
#part_2.
def mean(arr):
  sum=0
  for i in arr:
    sum+=i

  return sum/len(arr)

def sd(arr):
  tot=0
  for i in arr:
    tot+=(i-mean(arr))**2
  return (tot/(len(arr)-1))**0.5

def new(arr):
  count=0
  c=sd(arr)*1.5
  positive_c=mean(arr)+c
  negative_c=mean(arr)-c

  for i in range(len(arr)):
    if arr[i]>positive_c or arr[i]<negative_c:
      count+=1
  l=np.array([0]*count)
  j=0
  for i in range(len(arr)):
    if arr[i]>positive_c or arr[i]<negative_c:
      l[j]=arr[i]
      j+=1

  return l


arr=[10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]

print("The mean of the numbers is:",mean(arr))
print("The standard deviation is:",sd(arr))
print("New array:",new(arr))
```
